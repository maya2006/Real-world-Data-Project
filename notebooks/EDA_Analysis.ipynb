```python
# ── Import Libraries ──
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
print('Libraries loaded successfully ✓')
```

```python
# ── Load Dataset ──
df = pd.read_csv('../dataset/diabetes.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# ── Dataset Info ──
df.info()
```

```python
# ── Statistical Summary ──
df.describe().round(2)
```

```python
# ── Missing Values ──
print('Missing values per column:')
print(df.isnull().sum())
print(f'\nTotal zeros in Glucose: {(df["Glucose"]==0).sum()}')
print(f'Total zeros in BMI:     {(df["BMI"]==0).sum()}')
```

```python
# ── Outcome Distribution ──
fig, ax = plt.subplots()
outcome_counts = df['Outcome'].value_counts()
ax.bar(['Non-Diabetic (0)', 'Diabetic (1)'],
       outcome_counts.values, color=['#4C9BE8', '#E8544C'])
ax.set_title('Outcome Distribution', fontsize=14, fontweight='bold')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# ── Feature Histograms ──
df.hist(bins=20, figsize=(16, 10), color='#4C9BE8', edgecolor='white')
plt.suptitle('Feature Distributions', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```

```python
# ── Correlation Heatmap ──
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn',
            linewidths=0.5, square=True)
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# ── Glucose Analysis ──
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, (outcome, color, label) in zip(axes, [(0,'#4C9BE8','Non-Diabetic'), (1,'#E8544C','Diabetic')]):
    data = df[df['Outcome']==outcome]['Glucose']
    ax.hist(data, bins=20, color=color, alpha=0.8, edgecolor='white')
    ax.set_title(f'Glucose — {label}', fontweight='bold')
    ax.set_xlabel('Glucose')
    ax.set_ylabel('Frequency')
    ax.axvline(data.mean(), color='red', linestyle='--', label=f'Mean: {data.mean():.1f}')
    ax.legend()
plt.suptitle('Glucose Distribution by Outcome', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── BMI Analysis ──
fig, ax = plt.subplots(figsize=(7, 5))
df.boxplot(column='BMI', by='Outcome', ax=ax)
ax.set_title('BMI Distribution by Outcome')
ax.set_xlabel('Outcome (0=Non-Diabetic, 1=Diabetic)')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# ── Pairplot (key features) ──
key_features = ['Glucose', 'BMI', 'Age', 'Insulin', 'Outcome']
sns.pairplot(df[key_features], hue='Outcome',
             palette={0:'#4C9BE8', 1:'#E8544C'},
             diag_kind='hist', plot_kws={'alpha':0.5})
plt.suptitle('Pairplot of Key Features', y=1.01)
plt.show()
```
